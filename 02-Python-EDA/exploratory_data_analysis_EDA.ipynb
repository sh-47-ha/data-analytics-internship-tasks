# Visualization

1. Bar Chart

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Category", data=df)

plt.title("Number of Orders by Category")
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(y="Sub-Category", data=df)

plt.title("Orders by Sub-Category")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Segment", data=df)

plt.title("Orders by Segment")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Region", data=df)

plt.title("Orders by Region")
plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.countplot(y="State", data=df)

plt.title("Orders by State")
plt.show()
```

2. Scatter Plot

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Sales", y="Profit", data=df)

plt.title("Sales vs Profit")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Discount", y="Profit", data=df)

plt.title("Discount vs Profit")
plt.show()
```

3. Histogram

```python
plt.figure(figsize=(7,5))

sns.histplot(df["Sales"], bins=30)

plt.title("Distribution of Sales")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(df["Profit"], bins=30)

plt.title("Distribution of Profit")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(df["Quantity"], bins=10)

plt.title("Distribution of Quantity")
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.histplot(df["Discount"], bins=10)

plt.title("Distribution of Discount")
plt.show()
```

4. Heatmap

```python
corr = df[["Sales", "Profit", "Discount", "Quantity"]].corr()
plt.figure(figsize=(6,4))

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```

# Outlier Detection
